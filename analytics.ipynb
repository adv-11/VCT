```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('datasets/vct-champions.csv')
df
```

```python
top_players_fkpr = df.nlargest(10, 'fkpr')

sns.set_theme(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.scatterplot(data=top_players_fkpr, x='fkpr', y='fdpr', hue='team', style='team', s=100)

plt.title('Top 10 Players: FKPR (First Kills per Round) vs FDPR (First Deaths per Round)', fontsize=16)
plt.xlabel('First Kills per Round (FKPR)', fontsize=12)
plt.ylabel('First Deaths per Round (FDPR)', fontsize=12)

plt.legend(title='Team')
plt.show()
```

```python
top_players_fkpr = df.nlargest(10, 'fkpr')

sns.set_theme(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.scatterplot(data=top_players_fkpr, x='fkpr', y='fdpr', hue='player', style='player', s=100)

plt.title('Top 10 Players: FKPR (First Kills per Round) vs FDPR (First Deaths per Round)', fontsize=16)
plt.xlabel('First Kills per Round (FKPR)', fontsize=12)
plt.ylabel('First Deaths per Round (FDPR)', fontsize=12)

plt.legend(title='Team')
plt.show()
```

```python
average_acs_per_team = df.groupby('team')['acs'].mean().reset_index()

sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.barplot(data=average_acs_per_team, x='team', y='acs', hue='acs')

plt.title('Average ACS (Average Combat Score) per Team', fontsize=16)
plt.xlabel('Team', fontsize=12)
plt.ylabel('Average ACS', fontsize=12)

plt.xticks(rotation=45)
plt.show()
```

```python
top_players_kast = df.nlargest(10, 'kast')

sns.set(style="whitegrid")

plt.figure(figsize=(12, 8))
bar_plot = sns.barplot(data=top_players_kast, x='kast', y='player', hue='player', palette='viridis')

plt.title('Top 10 Players: KAST Consistency (Kill, Assist, Survive, Trade)', fontsize=16)
plt.xlabel('KAST Percentage', fontsize=12)
plt.ylabel('Player', fontsize=12)

for index, value in enumerate(top_players_kast['kast']):
    bar_plot.text(value + 0.01, index, f'{value:.2f}', color='black', va='center')

plt.legend(title='Player', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
top_players_acs = df.nlargest(10, 'acs')

sns.set(style="whitegrid")

plt.figure(figsize=(12, 8))
sns.scatterplot(data=top_players_acs, x='adr', y='acs', hue='player', style='team', s=200)


plt.title('Top 10 Players: ADR (Average Damage per Round) vs ACS (Average Combat Score)', fontsize=16)
plt.xlabel('Average Damage per Round (ADR)', fontsize=12)
plt.ylabel('Average Combat Score (ACS)', fontsize=12)

plt.legend(title='Player', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
top_players_kmax = df.nlargest(10, 'kmax')

# Set the style for seaborn
sns.set(style="whitegrid")

# Create a bar plot for KMAX for top 10 players
plt.figure(figsize=(12, 6))
sns.barplot(data=top_players_kmax, x='player', y='kmax', palette='viridis')

# Add titles and labels
plt.title('Top 10 Players: Max Kills in a Map (KMAX)', fontsize=16)
plt.xlabel('Player', fontsize=12)
plt.ylabel('Max Kills in a Map (KMAX)', fontsize=12)

# Add data labels on top of bars
for index, value in enumerate(top_players_kmax['kmax']):
    plt.text(index, value + 0.1, f'{value}', color='black', ha='center')

# Show plot
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='team', y='kd', palette='viridis')
plt.title('Distribution of K/D Ratios by Team', fontsize=16)
plt.xlabel('Team', fontsize=12)
plt.ylabel('K/D Ratio', fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.violinplot(data=df, x='team', y='hs', palette='viridis')
plt.title('Distribution of Headshot Percentages by Team', fontsize=16)
plt.xlabel('Team', fontsize=12)
plt.ylabel('Headshot Percentage', fontsize=12)
plt.show()
```

```python
top_players_clutch = df.nlargest(10, 'cls')


clutch_data = top_players_clutch.pivot_table(values='cls', index='player', columns='team', aggfunc='mean')

plt.figure(figsize=(12, 8))

sns.heatmap(clutch_data, annot=True, cmap='viridis', cbar_kws={'label': 'Average Clutches Survived'})

plt.title('Average Clutches Survived by Player and Team', fontsize=16)
plt.xlabel('Team', fontsize=12)
plt.ylabel('Player', fontsize=12)

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.kdeplot(data=df, x='adr', hue='team', fill=True, common_norm=False, palette='viridis')
plt.title('Kernel Density Estimation of Average Damage per Round (ADR)', fontsize=16)
plt.xlabel('Average Damage per Round (ADR)', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.show()
```

```python
# 1. Highest Number of Rounds Played by a Team
highest_rounds_team = df.groupby('team')['rnd'].sum().idxmax()
highest_rounds_value = df.groupby('team')['rnd'].sum().max()

print(f"Highest Number of Rounds Played: {highest_rounds_value} by Team: {highest_rounds_team}")

# 2. Highest Average Combat Score of Players with More Than 300 Matches
players_over_300 = df[df['rnd'] > 300]
highest_acs_player = players_over_300.loc[players_over_300['acs'].idxmax()]

print(f"Highest Average Combat Score (ACS): {highest_acs_player['acs']} by Player: {highest_acs_player['player']}")

# 3. Maximum Kills in a Map (KMAX)
max_kills_player = df.loc[df['kmax'].idxmax()]

print(f"Max Kills in a Map (KMAX): {max_kills_player['kmax']} by Player: {max_kills_player['player']}")
```
